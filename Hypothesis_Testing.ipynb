# Hypothesis:

**H0: Distribution of tip for yellow taxi = Distribution of tip for green taxi**

**H1: Distribution of tip for yellow taxi â‰  Distribution of tip for green taxi**

```python
# Independent T-Test
from scipy import stats
yellow_taxi_tip = df[df['color']=='yellow']['tip']
green_taxi_tip = df[df['color']=='green']['tip']
t_stat, p_val = stats.ttest_ind(yellow_taxi_tip, green_taxi_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
if p_val < 0.05:
    print("Reject null hypothesis: Tip differs by Color")
else:
    print("Fail to reject null hypothesis: No significant tip difference by Color")
```

# Hypothesis:

**H0: Distribution of tip for Cash Payment = Distribution of tip for Card Payment**

**H1: Distribution of tip for Cash Payment â‰  Distribution of tip for Card Payment**

```python
# Independent T-Test
from scipy import stats
credit_card_pay_tip = df[df['payment']=='credit card']['tip']
cash_pay_tip = df[df['payment']=='cash']['tip']
t_stat, p_val = stats.ttest_ind(credit_card_pay_tip, cash_pay_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
if p_val < 0.05:
    print("Reject null hypothesis: Tip differs by Payment Method")
else:
    print("Fail to reject null hypothesis: No significant tip difference by Payment Method")
```

# Hypothesis:

**H0: Distribution of tip is the same across all pickup_borough categories.**

**H1: At least one category has a different distribution of tip.**

```python
# ANOVA Test
manhattan_tips=df[df['pickup_borough']=='Manhattan']['tip']
brooklyn_tips=df[df['pickup_borough']=='Brooklyn']['tip']
queens_tips=df[df['pickup_borough']=='Queens']['tip']
bronx_tips=df[df['pickup_borough']=='Bronx']['tip']
f_stat, p_val = stats.f_oneway(manhattan_tips, brooklyn_tips, queens_tips, bronx_tips)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Tip differs by Pickup Borough")
else:
    print("Fail to reject null hypothesis: No significant difference in tip by Pickup Borough")
```

# Hypotheses for Tukey test (for each pair of groups):

**H0: Distribution of group1 = Distribution of group2**

**H1: Distribution of group1 â‰  Distribution of group2**

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=df.tip, groups=df.pickup_borough, alpha=0.05)
print(tukey_result)
```

# Training Score

```python
y_train_pred=lr.predict(x_train)
train_score=lr.score(x_train,y_train)
print("Model Training Score : ",train_score)
```

```python
y_test_pred=lr.predict(x_test)
test_score=lr.score(x_test,y_test)
print("Model Testing Score : ",test_score)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_squared_error,mean_absolute_error,r2_score
rmse_train=np.sqrt(root_mean_squared_error(y_train,y_train_pred))
print("RMSE for training data : ",rmse_train)
rmse_test=np.sqrt(root_mean_squared_error(y_test,y_test_pred))
print("RMSE for testing data : ",rmse_test)
mse_train=mean_squared_error(y_train,y_train_pred)
print("MSE for training data : ",mse_train)
mse_test=mean_squared_error(y_test,y_test_pred)
print("MSE for testing data : ",mse_test)
mae_train=mean_absolute_error(y_train,y_train_pred)
print("MAE for training data : ",mae_train)
mae_test=mean_absolute_error(y_test,y_test_pred)
print("MAE for testing data : ",mae_test)
r2_train=r2_score(y_train,y_train_pred)
print("R2 Score for training data : ",r2_train)
r2_test=r2_score(y_test,y_test_pred)
print("R2 Score for testing data : ",r2_test)
```
