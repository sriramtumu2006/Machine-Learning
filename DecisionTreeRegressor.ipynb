```python
import numpy as np;
import matplotlib.pyplot as plt;
# Units consumed
x=np.array([50,120,180,250,320,380]).reshape((-1,1))
# Corresponding bills
y=np.array([500,1000,1500,1800,2500,2500])
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg=DecisionTreeRegressor(max_depth=3,random_state=42)
dt_reg.fit(x,y)
```

```python
x_grid=np.linspace(0,400,200).reshape((-1,1))
y_pred=dt_reg.predict(x_grid)
plt.scatter(x,y,color='blue')
plt.plot(x_grid,y_pred,color='red')
plt.xlabel("Units Consumed")
plt.ylabel("Electricity Bill")
plt.title("Decision Tree Regression - Electricity Bill Slabs")
plt.show()
```

```python
units=float(input("Enter electricity units consumed: "))
bill=dt_reg.predict([[units]])
print("Estimated electricity bill:",bill[0])
```

# Predict Salary based on years of experience

```python
x=np.array([1,2,3,4,5,7,8,9]).reshape((-1,1))
y=np.array([3,3.5,4,4.2,6,6.5,7,7.2])
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
dt_reg=DecisionTreeRegressor(random_state=42)
dt_reg.fit(x_train,y_train)
```

```python
x_grid=np.linspace(1,8,100).reshape((-1,1))
y_pred=dt_reg.predict(x_grid)
plt.scatter(x,y,color='blue')
plt.plot(x_grid,y_pred,color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Decision Tree Regression - Salary Prediction")
plt.show()
```

```python
dt_deep=DecisionTreeRegressor(max_depth=None)
dt_deep.fit(x_train,y_train)
y_pred_deep=dt_deep.predict(x_grid)
plt.scatter(x,y)
plt.plot(x_grid,y_pred_deep,color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")    
plt.title("Decision Tree Regression (Deep) - Salary Prediction")
plt.show()
```

```python
dt_controlled=DecisionTreeRegressor(max_depth=3)
dt_controlled.fit(x_train,y_train)
y_pred_controlled=dt_controlled.predict(x_grid)
plt.scatter(x,y)
plt.plot(x_grid,y_pred_controlled,color='red')
plt.title("Decision Tree Regression (Controlled) - Salary Prediction")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_test_pred_deep=dt_deep.predict(x_test)
y_test_pred_controlled=dt_controlled.predict(x_test)
print("Deep Tree Model Performance:")
print("Mean Squared Error:", mean_squared_error(y_test,y_test_pred_deep))
print("R2 Score:", r2_score(y_test,y_test_pred_deep))
print("Controlled Tree Model Performance:")
print("Mean Squared Error:", mean_squared_error(y_test,y_test_pred_controlled))
print("R2 Score:", r2_score(y_test,y_test_pred_controlled))
```
